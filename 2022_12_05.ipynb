```python
with open('input/2022_12_05_input') as file:
    f = file.read()
```

```python
towers = f.split('\n\n')[0].replace(']     [', '] [x] [').replace(
    ']         [','] [x] [x] [').replace('           ', ' [x] [x] [x]').replace('    ', ' [x] ')

orders = f.split('\n\n')[1].strip().split('\n')

matrix = []
for row in towers.split('\n')[:-1]:
    matrix.append(row.split(' '))
    
import pandas as pd
df = pd.DataFrame(matrix).drop(9,axis=1) 
df
```

```python
import numpy as np
df_empty = pd.DataFrame('[x]', index=np.arange(0, df.shape[0]*df.shape[1] - df.shape[1]), columns=np.arange(0,9,1))
df_final = pd.concat([df_empty,df])
df_final = df_final.reset_index(drop=True)
df_final.columns = np.arange(1,10)
df_final
```

```python
for order in orders:
    number = int(order.split()[1])
    from_ = int(order.split()[3])
    to_ = int(order.split()[5])
    
    last_i_from = df_final[from_].where(df_final[from_]=='[x]').last_valid_index() + 1
    number_list = list(df_final[from_][last_i_from: last_i_from + number])
    first_i_to = df_final[to_].where(df_final[to_]=='[x]').last_valid_index() 
    #if not first_i_to:
    
    for i in range(0, len(number_list)):
        df_final.loc[first_i_to - i, to_] = number_list[i]
        df_final.loc[last_i_from + i, from_] = '[x]'
```

```python
df_final
```

```python
sol = ''
for col in df_final.columns:
	sol = sol + df_final[col][df_final[col].where(df_final[col]=='[x]').last_valid_index() + 1]
    
print(sol.replace('[', '').replace(']',''))
```
