# Challenge 1

```python
with open('input/2022_12_01_input') as file:
    f = file.read()
in_list = f.split('\n\n')
sublists = [x.split() for x in in_list]
sublists_int = [[int(x) for x in lst] for lst in sublists]
sums = [sum(i) for i in sublists_int]
print(max(sums))
```
