# Challenge 1

```python
with open('input/2022_12_03_input') as file:
    f = file.read()
    
rucksacks = f.split()
sums = 0
for rucksack in rucksacks:
    first_part = rucksack[0:int(len(rucksack)/2)]
    c = 0
    for item in first_part:
        if item in rucksack[int(len(rucksack)/2):]:
            if item.islower():
                sums = sums + (ord(item) - ord('a') + 1)
            else:
                sums = sums + (ord(item) - ord('A') + 27)                
            break
            
print(sums)
```

```python
with open('input/2022_12_03_input') as file:
    f = file.read()
    
rucksacks = f.split()

sums_alt = 0
items = [list(set(rucksack[:int(len(rucksack)/2)]) & set(rucksack[int(len(rucksack)/2):]))[0] for rucksack in rucksacks]
for item in items:
    if item.islower():
        sums_alt += (ord(item) - ord('a') + 1)
    else:
        sums_alt += (ord(item) - ord('A') + 27) 

print(sums_alt)
```
