# Challenge 1

```python
with open('input/2022_12_04_input') as file:
    f = file.read()
    
pairs =f.split()
count = 0
for pair in pairs:
    splitted = pair.split(',')
    first = range(int(splitted[0].split('-')[0]), int(splitted[0].split('-')[1]) , 1)
    second = range(int(splitted[1].split('-')[0]), int(splitted[1].split('-')[1]), 1)
    if (first.start >= second.start and first.stop <= second.stop) or (first.start <= second.start and first.stop >= second.stop):
        count = count + 1
        
print(count)    
```

# Challenge 2

```python
count2 = 0
for pair in pairs:
    splitted = pair.split(',')
    first = range(int(splitted[0].split('-')[0]), int(splitted[0].split('-')[1]) + 1, 1)
    second = range(int(splitted[1].split('-')[0]), int(splitted[1].split('-')[1]) + 1, 1)
    if len(set(first) & set(second)) != 0:
        count2 = count2 + 1
        
print(count2)  
```
