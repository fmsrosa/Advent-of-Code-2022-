```python
with open('input/2022_12_09_input.txt') as file:
    f = file.read()
    
#f = '''R 5
#U 8
#L 8
#D 3
#R 17
#D 10
#L 25
#U 20'''  
    
commands = f.splitlines()
mat = np.zeros((7000, 7000), dtype=np.int)

current_head_row = 3500#len(matrix)-1
current_head_col = 3500

current_tail_row = 3500#len(matrix)-1
current_tail_col = 3500

_list = [[3500,3500] for _ in range(10)]


c = 0
for command in commands:
    #print(command)
    direction = command.split()[0]
    value = int(command.split()[1])
    #print('value', value)
    for val in range(value):
        #print('before', value, direction, current_head_row, current_head_col, current_tail_row, current_tail_col)
        if direction == 'R':
            _list[0][1] += 1
        elif direction == 'L':
            _list[0][1] -= 1
        elif direction == 'U':
            _list[0][0] -= 1
        elif direction == 'D':
            _list[0][0] += 1
        for i in range(9):
            current_head_col = _list[i][1]
            current_head_row = _list[i][0]
            current_tail_col = _list[i+1][1]
            current_tail_row = _list[i+1][0]
            #print('quasi', direction, current_head_row, current_head_col, current_tail_row, current_tail_col)    
            if current_head_col == current_tail_col:
                if current_head_row > current_tail_row:
                    current_tail_row += abs(current_head_row - current_tail_row)//2
                else:
                    current_tail_row -= abs(current_head_row - current_tail_row)//2
            elif current_head_row == current_tail_row:
                if current_head_col > current_tail_col:
                    current_tail_col += abs(current_head_col - current_tail_col)//2
                else:
                    current_tail_col -= abs(current_head_col - current_tail_col)//2
            elif np.linalg.norm(np.array((current_tail_row,current_tail_col)) - np.array((current_head_row,current_head_col))) > 1.5:
                    current_tail_row += abs(current_head_row - current_tail_row)//(current_head_row - current_tail_row)
                    current_tail_col += abs(current_head_col - current_tail_col)//(current_head_col - current_tail_col)
                    
            _list[i+1][0] = current_tail_row
            _list[i+1][1] = current_tail_col

            #if c > 1900:
            #	print(direction, value, current_head_row, current_head_col, current_tail_row, current_tail_col)
            mat[_list[-1][0], _list[-1][1]] = 1
            #print(command)  
            #print(mat)
        #print(command, _list)
```
