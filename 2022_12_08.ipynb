```python
with open('input/2022_12_08_input') as file:
    f = file.read()
    
f_test = '''
30373
25512
65332
33549
35390
'''
        
clean =','.join( f.strip())
#matrix = np.matrix(clean)

df = pd.DataFrame([x.split(',') for x in clean.split('\n,')])
df = df.drop(df.shape[1]-1, axis=1)
```

```python
non_visible = 0
for i in range(1, df.shape[0]-1):
    for j in range(1, df.shape[1]-1):
        max_left = df.loc[i,0:j-1].max()
        max_right = df.loc[i,j+1:df.shape[1]].max()
        max_top = df.loc[0:i-1,j].max()
        max_bottom = df.loc[i+1:df.shape[0],j].max()
        #print('b',i, j, df.loc[i,j], max_top, max_bottom, max_left, max_right)
        if df.loc[i,j] <= max_top and df.loc[i,j] <= max_bottom and df.loc[i,j] <= max_left and df.loc[i,j] <= max_right:
        #    print(i, j, df.loc[i,j], max_top, max_bottom, max_left, max_right)
            non_visible += 1
            
visible_all = df.shape[0] * df.shape[1] - non_visible
            
print(visible_all)
```

```python
highest = 0
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
#for i in range(3,4):
#    for j in range(2,3):
        val = df.loc[i,j]  
        left = 0
        right = 0
        top = 0
        bottom = 0
        for l in range(j-1, -1, -1):
           # print('l', l)
            if df.loc[i,l] < val:
          #      print('left', l)
                left += 1
            else:
                if df.loc[i,l] >= val:
                    left += 1
                break
        for r in range(j+1, df.shape[1]):
            if df.loc[i,r] < val:
         #       print('right', r)
                right += 1
            else:
                if df.loc[i,r] >= val:
                    right += 1
                break
        for t in range(i-1, -1, -1):
            if df.loc[t,j] < val:
        #        print('top', t)
                top += 1
            else:
                if df.loc[t,j] >= val:
                    top += 1
                break   
        for b in range(i+1, df.shape[0]):
            if df.loc[b,j] < val:
         #       print('bottom', b)
                bottom += 1
            else:
                if df.loc[b,j] >= val:
                    bottom += 1
                break 
        #print(left , right , top , bottom)
        tmp = left * right * top * bottom
        #print(i,j, tmp)
        if tmp > highest:
            highest = tmp
            
print(highest)
```
