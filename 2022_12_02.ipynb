# Challenge 1

```python
import pandas as pd
```

```python
df = pd.read_csv('input/2022_12_02_input', sep=' ', names=['Opponent', 'Me'])

df['shape_score'] = df['Me'].map({'X': 1, 'Y': 2, 'Z': 3})
        
df['opponent_tmp'] = df['Opponent'].map({'A': -1, 'B': -2, 'C': -3})
df['outcome_score'] = (df['opponent_tmp'] + df['shape_score']).map(lambda row: 3 if row == 0
                                                                   else(6 if row == 1 or row == -2
                                                                        else 0)
                                                                  )

df['score'] = df['outcome_score'] + df['shape_score']

df['score'].sum()
```

# Challenge 2

```python
df['outcome_score_2'] = df['Me'].map({'X': 0, 'Y': 3, 'Z': 6})

df['shape_victory'] = (df['opponent_tmp'] * -1).map(lambda row: row+1 if row!=3 else 1)
df['shape_draw'] = (df['opponent_tmp'] * -1)
df['shape_loss'] = df.apply(lambda row: list(set([1,2,3]) - set([row['shape_victory'], row['shape_draw']]))[0], axis=1)

df['shape_played_me'] = df.apply(lambda row: row['shape_victory'] if row['outcome_score_2'] == 6 
                                     else (row['shape_draw'] if row['outcome_score_2'] == 3
                                           else row['shape_loss']), axis=1)

df['score2'] = df['outcome_score_2'] + df['shape_played_me']

df['score2'].sum()
```
